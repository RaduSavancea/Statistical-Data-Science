```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
#Aufgabe 3.2
mean = [0, 0, 0]
cov = [[1, 0, 0], [0, 1, 0], [0, 0, 1]] 
x = np.random.multivariate_normal(mean, cov, 1000)
x_new=np.zeros(shape=(1000,1))
y_new=np.zeros(shape=(1000,1))
z_new=np.zeros(shape=(1000,1))
```

```python
for i in range((x.shape[0])):
    x[i]= x[i]/np.linalg.norm(x[i])
    x_new[i]=x[i][0]
    y_new[i]=x[i][1]
    z_new[i]=x[i][2]
```

```python
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
c = np.arange(len(x)) / len(x)  
p = ax.scatter(x_new, y_new, z_new, c=0.5*c, cmap=plt.cm.magma)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_box_aspect([np.ptp(i) for i in x.T])  
fig.colorbar(p, ax=ax)
plt.show()
# Magma sphere  
```

```python
#Aufgabe 3.3
x_uni=np.random.uniform(low=0.0, high=2*np.pi, size=1000)
y_uni=np.random.uniform(low=0.0, high=np.pi, size=1000)
```

```python
def func1(x, y):
    return np.sin(y)*np.cos(x)
def func2(x, y):
    return np.sin(x)*np.sin(y)
def func3(y):
    return np.cos(y)
```

```python
x_new_sp=np.zeros(shape=(1000,1))
y_new_sp=np.zeros(shape=(1000,1))
z_new_sp=np.zeros(shape=(1000,1))

for i in range((x.shape[0])):
    x_new_sp[i]= func1(x_uni[i], y_uni[i])
    y_new_sp[i]= func2(x_uni[i], y_uni[i])
    z_new_sp[i]= func3( y_uni[i])
```

```python
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
c = np.arange(len(x)) / len(x)  
p = ax.scatter(x_new_sp, y_new_sp, z_new_sp, c=0.5*c, cmap=plt.cm.magma)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_box_aspect([np.ptp(i) for i in x.T])  

fig.colorbar(p, ax=ax)
plt.show()
```

The points don't appear to be uniformly distributed, since we can clearly see that they cluster on the upper and lower side of the magma sphere, this being the major difference between the 2 plots.
