```python
import matplotlib.pyplot as plt
import numpy as np
import csv
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
weight=[]
bidia=[]
abdcir=[]
with open('birthweight.csv', newline='') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            weight.append(float(row["weight"]))
            bidia.append(float(row["bidia"]))
            abdcir.append(float(row["abdcir"]))
a=np.linspace(51,90,num=40)
a2=np.linspace(480,2380,num=1901)
a3=np.linspace(16.4,29.3,num=130)
```

Interprete results: The distribution is left skewed. Most of the babies have a head diameter from 70-80 and because the distribution is left skewed it means that we have more babies with larger head diameter than babies with smaller head diameter.

1.2)
$F(x,y) = 1/n *\sum_{i=1}^n 1_{x_i<=x,y_i<=y}$ 

```python
def c(x,y,arr):
    count=0
    for i in range(len(bidia)):
        if(arr[i]<=x and weight[i]<=y):
            count=count+1
    return count*1/50


test1=np.ones((len(a),len(a2)))
test1=test1.astype(tuple)
for i in range(len(a)):
    for j in range(len(a2)):
        test1[i][j]=(a[i],a2[j])

p1=np.ones((len(a),len(a2)))
for i in range(len(a)):
    for j in range(len(a2)):
        p1[i,j]=c(test1[i,j][0],test1[i,j][1],bidia)

X, Y = np.meshgrid(a, a2)
p1=np.transpose(p1)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.set_title("bidia/weight")
ax.plot_surface(X, Y, p1,cmap="plasma")
plt.show()
```
