```python
import matplotlib.pyplot as plt
import numpy as np
import csv
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
eruptions=[]
waiting=[]
with open('faithful.csv', newline='') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            eruptions.append(float(row["eruptions"]))
            waiting.append(float(row["waiting"]))
```

```python
m=40
s=10
def count(arr , k):
    tmp=0
    for i in range(len(arr)):
        if(arr[i]==k):
            tmp=tmp+1
    return tmp/len(waiting)

x=np.linspace(1,s**2-1,s**2-1)
y=[]
fig, axs = plt.subplots(s, s,figsize=(22,22))
for j in range(1,s**2):
    dist=60.0/j
    h=np.zeros(len(waiting))
    for i in range(len(waiting)):
        for k in range(1,j+1):
            if(waiting[i]<dist*k+m):
                h[i]=k
                break
    ey=0
    for i in range(1,j+1):
        ey=ey+(i-1/2)*(count(h,i))
    y.append(m+(dist*ey))
    c1=j//s
    c2=j%s
    axs[c1, c2].hist(h,bins=j)
    #print(h)
    #print(max(h))
```

```python
j=3
dist=(100-40)/j
h=np.zeros(len(waiting))
for i in range(len(waiting)):
    for k in range(1,j+1):
        if(waiting[i]<dist*k+m):
            h[i]=k
            break
plt.hist(h,bins=j)
plt.xlabel('Waiting Bin Repartition')
plt.ylabel('Bin Density')
plt.show()
```

```python
j=16
dist=(100-40)/j
h=np.zeros(len(waiting))
for i in range(len(waiting)):
    for k in range(1,j+1):
        if(waiting[i]<dist*k+m):
            h[i]=k
            break
plt.hist(h,bins=j)
plt.xlabel('Waiting Bin Repartition')
plt.ylabel('Bin Density')
plt.show()
```

For k=3 we can observe the unimodal histogram and for k=16 the bimodal one.
A better representation of the data is given by the second plot(k=16), beacause for a low number of bins we can't get a feeling on how the data is really distributed.

```python
plt.plot(x,y)
mean=sum(y)/len(y)
plt.plot([0,100],[mean,mean])
plt.xlabel('Bin Repartition')
plt.ylabel('Histogram mean')
plt.show()
```
